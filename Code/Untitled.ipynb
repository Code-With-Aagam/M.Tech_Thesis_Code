```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Input range
x = np.linspace(-6, 6, 1000)

# Compute outputs
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y_sigmoid, label="Sigmoid", linewidth=3)
plt.plot(x, y_tanh, label="Tanh", linewidth=3)
plt.plot(x, y_relu, label="ReLU", linewidth=3)

# Saturation lines
plt.axhline(y=1, color="gray", linestyle="--", alpha=0.6)
plt.axhline(y=-1, color="gray", linestyle="--", alpha=0.6)
plt.axhline(y=0, color="gray", linestyle="--", alpha=0.6)

# Annotations (bigger font)
plt.annotate("Sigmoid saturation → 0", xy=(-5, 0), xytext=(-4.5, 0.25),
             fontsize=12, arrowprops=dict(facecolor="black", arrowstyle="->"))

plt.annotate("Tanh saturation → -1", xy=(-5, -1), xytext=(-4.5, -1.3),
             fontsize=12, arrowprops=dict(facecolor="black", arrowstyle="->"))

plt.annotate("Tanh saturation → 1", xy=(5, 1), xytext=(2.5, 1.25),
             fontsize=12, arrowprops=dict(facecolor="black", arrowstyle="->"))

plt.annotate("Sigmoid saturation → 1", xy=(5, 1), xytext=(2.5, 1.05),
             fontsize=12, arrowprops=dict(facecolor="black", arrowstyle="->"))

# Labels and legend (bigger font)
plt.title("Activation Functions with Saturation Zones", fontsize=18)
plt.xlabel("Input", fontsize=15)
plt.ylabel("Output", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ylim(-1.5, 1.5)
plt.legend(fontsize=15)
plt.grid(True)
plt.savefig('AF')
plt.show()
```
